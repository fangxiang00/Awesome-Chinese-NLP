```python
"""如何声明一个类"""

#类的组成：类属性、实例方法、静态方法、类方法
class Student:
  native_place='江苏' # 类属性（直接写在类里的变量）
  
  # 初始化方法
  def __init__(self,name,age):
    # self.name为实例属性,进行了一个赋值的操作，将局部变量name的值赋值实例属性
    self.name = name
    self.age = age

  # 实例方法（在类外定义的函数称为函数，类内定义称为方法）
  def info(self):
    print('我的名字叫：',self.name,'年龄是:',self.age)
  
  def eat(self):
    print("学生在吃饭")

  #类方法：使用类名直接访问的方法
  @classmethod
  def cm(cls):
    print('我是类方法')
  
  #静态方法（不允许写self），也是使用类名直接访问
  @staticmethod
  def sm(): #没有默认参数
    print('静态方法')
```

## 封装

```python
class Student:
  def __init__(self, age):
    self.set_age(age)
  def get_age(self):
    return self.__age
  def set_age(self,age):
    if 0<=age<=120:
      self.__age=age
    else:
      self.__age=18
stu1 = Student(130)
stu2 = Student(30)
stu1.get_age(),stu2.get_age()
```

```python
class Student:
  def __init__(self,name,age):
    self.name = name
    self.__age = age
  def show(self):
    print(self.name, self.__age)
stu = Student('张三',20)
stu.show()
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def info(self):
    print("姓名：{0}，年龄：{1}".format(self.name, self.age))
```

```python
class Student(Person):
  def __init__(self, score,name,age):
    super().__init__(name,age) #继承父类的init方法
    self.score = score
```

```python
stu = Student(30,'fangxiang',30)
```

```python
stu.name
```

super(Net, self).__init__()的作用：


首先找到Net的父类（比如是类NNet），然后把类Net的对象self转换为类NNet的对象，然后“被转换”的类NNet对象调用自己的init函数

这是对继承自父类的属性进行初始化。而且是用父类的初始化方法来初始化继承的属性。

也就是说，子类继承了父类的所有属性和方法，父类属性自然会用父类方法来进行初始化。

```python
class mywindow(QtWidgets.QMainWindow, Ui_MainWindow):
    def __init__(self):
        super(mywindow, self).__init__()
        self.setupUi(self)
 
    # 定义槽函数
    def hello(self):
        self.textEdit.setText("hello world")
```
